### Download Tabular Patient Data

```python
# Data source: https://physionet.org/content/mimiciii-demo/1.4/
# Note: make sure to copy the right link
download_url = 'https://physionet.org/files/mimiciii-demo/1.4/PRESCRIPTIONS.csv?download'
```

```python
filename = download_url.split('/')[-1].split('?')[0]
print(filename)

folder = './data/'
```

```python
# Download from URL and save to disk
response = requests.get(download_url)

with open(os.path.join(folder,filename), 'wb') as f:
    f.write(response.content)
```

```python
# read from this
df = pd.read_csv(folder+filename)
```
