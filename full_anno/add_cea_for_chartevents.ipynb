```python
import pandas as pd
import numpy as np
name = "CHARTEVENTS_anno.xlsx"
data = pd.read_excel(name)
data
```

```python
colnames = data.columns
cols = colnames[range(1,len(colnames),2)]
cols_cea = [f"{el}_cea" for el in cols]
cols_cea
cur_cols = list(colnames[0:1])
for i in range(len(cols)):
    cur_cols.append(cols[i])
    cur_cols.append(cols_cea[i])
data.columns = cur_cols
data
```

```python
from collections import defaultdict
from shervin2_dict import label_dict, category_dict, param_dict, valset_dict

dict_annos = {'row_id': defaultdict(lambda: "https://browser.ihtsdotools.org/?perspective=full&conceptId1=118522005&edition=MAIN/2024-06-01&release=&languages=en"),
             'subject_id': defaultdict(lambda: "https://browser.ihtsdotools.org/?perspective=full&conceptId1=125676002&edition=MAIN/2024-06-01&release=&languages=en"),
             'hadm_id': defaultdict(lambda: "https://www.biomedit.ch/rdf/sphn-schema/sphn#AdministrativeCase"),
             'icustay_id': defaultdict(lambda: "https://browser.ihtsdotools.org/?perspective=full&conceptId1=309904001&edition=MAIN/2024-06-01&release=&languages=en"),
             'itemid': defaultdict(lambda: "https://browser.ihtsdotools.org/?perspective=full&conceptId1=118522005&edition=MAIN/2024-06-01&release=&languages=en"),
             'charttime': defaultdict(lambda: "https://www.w3.org/2001/XMLSchema#dateTime"),
             'storetime': defaultdict(lambda: "https://www.w3.org/2001/XMLSchema#dateTime"),
             'dbsource_D_ITEMS': defaultdict(lambda: "https://www.w3.org/2001/XMLSchema#string"),
             'cgid': defaultdict(lambda: "https://www.biomedit.ch/rdf/sphn-schema/sphn#Performer"),
             'valueuom': defaultdict(lambda: "https://build.fhir.org/valueset-ucum-units.xml"),
             'label_CAREGIVERS': defaultdict(lambda: "https://www.w3.org/2001/XMLSchema#string"),
             'description_CAREGIVERS': defaultdict(lambda: "https://www.w3.org/2001/XMLSchema#string"),
             'unitname_D_ITEMS': defaultdict(lambda: "https://build.fhir.org/valueset-ucum-units.xml"),
             'valueuom': defaultdict(lambda: "https://build.fhir.org/valueset-ucum-units.xml"),
             'label_D_ITEMS':label_dict,
             'abbreviation_D_ITEMS':label_dict,
              'category_D_ITEMS': category_dict, 
              'param_type_D_ITEMS': param_dict
             }

for col in dict_annos:
    for start in range(2,len(data)):
        if not pd.isnull(data.loc[start, f'{col}']):
            try:
                data.loc[start, f'{col}_cea'] = dict_annos[col][data.loc[start, f'{col}']]
            except KeyError:
                data.loc[start, f'{col}_cea'] = "https://www.w3.org/2001/XMLSchema#string"
data
```

```python
def is_float(element: any) -> bool:
    #If you expect None to be passed:
    if element is None: 
        return False
    try:
        float(element)
        return True
    except ValueError:
        return False
    
def is_integer(s):
    try:
        # Try to convert the string to a float
        num = float(s)
    except ValueError:
        # If conversion fails, it's not a number
        return False
    
    # Check if the float is an integer (has no fractional part)
    return num.is_integer()
```

```python
colval = 'value'
colnum = 'valuenum'
valset = set()
for start in range(2,len(data)):
    if not pd.isnull(data.loc[start, colval]):
        if is_float(data.loc[start, colval]):
            data.loc[start, f'{colval}_cea'] = "https://www.w3.org/2001/XMLSchema#double"
            if not pd.isnull(data.loc[start, colnum]):
                data.loc[start, f'{colnum}_cea'] = "https://www.w3.org/2001/XMLSchema#double"
        else:
            try:
                data.loc[start, f'{colval}_cea'] = valset_dict[data.loc[start, f'{colval}']]#"https://www.w3.org/2001/XMLSchema#string"
            except KeyError:
                data.loc[start, f'{colval}_cea'] = "https://www.w3.org/2001/XMLSchema#string"
            if not pd.isnull(data.loc[start, colnum]):
                if is_float(data.loc[start, colnum]):
                    data.loc[start, f'{colnum}_cea'] = "https://www.w3.org/2001/XMLSchema#double"
                else:
                    data.loc[start, f'{colnum}_cea'] = "https://www.w3.org/2001/XMLSchema#string"
            valset.add(data.loc[start, colval]) 
data
```
